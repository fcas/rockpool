# Training an spike-timing SNN to perform a supervised task

In this tutorial we will build a spiking network to solve a supervised learning task, based on copying an Arbitrary Dynamical System. We call these networks ADS networks for short. ADS networks are based on a body of research on efficient spike-time balanced networks from Sophie Denéve and Christian Machens [[1], [2], [3]]. This tutorial and implementation were developed by Julian Büchel.

[1]: https://doi.org/10.1371/journal.pcbi.1003258
[2]: https://doi.org/10.1038/nn.4243
[3]: https://arxiv.org/abs/1705.08026

ADS networks are able to learn to reproduce any non-linear dynamical system of the form

$$\dot{x} = f(x) + c$$

where $c$ is a continuous input and $f(\cdot)$ is an arbitrary non-linear function.

Being able to learn how to implement an arbitrary dynamical system in a spiking network is a powerful ability, which can be exploited to do complex tasks, such as training a classifier. In this tutorial we will tackle a temporal signal recognition task: spoken wake-phrase detection or keyword spotting.

## Background
<b>E</b>fficient <b>B</b>alanced <b>N</b>etworks (<b>EBN</b>s) [[2](https://www.nature.com/articles/nn.4243)] have coding properties that are of high interest to the neuromorphic community. They encode signals in an extremely precise and sparse way while still being robust to various types of noise. The network architecture is based on balanced inhibitory and excitatory recurrent connections, and works by internally maintaining a representation of some small number of dynamic state variables using a larger number of spiking neurons.

In 2017, Alemi et al. [[3](https://arxiv.org/abs/1705.08026)] demonstrated that a spiking network can be trained to follow the dynamics of a non-linear teacher system while maintaining the EBN network architecture, and the useful noise robustness properties that it brings.

In Rockpool we have adapted their learning rule and stripped down their architecture to the bare minimum of complexity, so we can build trained systems to perform signal processing tasks.

## Training approach
With the goal in mind to train a classifier using this network architecture, we recognise that a classic rate network with the following dynamics

$$\tau_j \dot{x}_j = -x_j + \hat{\mathbf{F}}c(t)_j + \hat{\mathbf{\Omega}}f(x(t))_j + b_j + \epsilon_j$$

actually implements a function of the form

$$\dot{x} = f(x) + c$$

which can be learned by the ADS Network! This means that we can train a rate network using BPTT (backpropagation through time) and then train a spiking ADS Network to mimic the rate networks dynamics.

This image shows the various networks and their configuration during learning for an audio task. The raw audio sample (at left) is filtered using a filterbank to produce input signal $c$, and passed as input to a recurrent rate network with $\hat{N}$ neurons. The rate network, consisting of feed-forward input weights $\hat{F}$, recurrent weights $\hat\Omega$ and readout decoder weights $\hat{\textbf{D}}$ is trained using BPTT (or any other desired training method), to produce a desired target dynamics $\hat{y}$ (green).

The rate network has now learned some useful recurrent dynamics, which we would like to mimic with a robust recurrent spiking network — the ADS network. The concept is that the ADS network will maintain an accurate representation of the $\hat{N}$ dynamical state variables from the rate network, as well as learn the dynamical system that the rate network embodies. We can then replace the recurrent rate network with the ADS spiking network, and the entire system should function as trained. 

```python
from IPython.display import Image
Image(filename='images/ads_learning_setup.png', width=800)
```

We now define a recurrent spiking network with $N$ neurons (where $N > 6\hat{N}$), with fast and slow recurrent weights $\Omega^f$ and $\Omega^s$ respectively. The fast recurrent weights are initialised to be in the efficient balanced regime, by choosing a random encoder matrix $\textbf{F}$ and analytically determining the matrix $\Omega^f$ according to the EBN network literature.

During training of the ADS network, an input signal $\textbf{c}$ is fed simultaneously into the rate network via $\hat{\textbf{F}}$ and into the ADS network via $\hat{\textbf{F}}\cdot\textbf{F}$. The ADS network adjusts the slow recurrent weights $\Omega^f$ to learn to mimic the rate dynamics $\hat{\textbf{x}}$, which it approximates by filtering the spikes to produce synaptic signals $\textbf{r}$ and then computing $\tilde{\textbf{x}} = \textbf{F}^T \textbf{r}$. The ADS network receives a continuous error signal $\textbf{e} = \hat{\textbf{x}} - \tilde{\textbf{x}}$ during training, which ensures that it remains close to the desired dynamics while learning them. This error signal is decayed during training and then switched off for inference.

```python
from IPython.display import Image
Image(filename='images/ads_inference_setup.png', width=800)
```

In inference mode, we simply throw away the recurrent rate network and replace it with $\textbf{F} \rightarrow$ ADS network $\rightarrow \textbf{F}^T$. We keep the trained input and decoder weights $\hat{\textbf{F}}$ and $\hat{\textbf{D}}$. The system now produces the trained output signal $\hat{\textbf{y}}$ using a spiking recurrent network.

# Training a temporal XOR task
We will train a rate network on a 1D temporal XOR task and then train an ADS network to copy the dynamics of this network.

Temporal XOR requires a network to detect and remember the polarity of two inputs that occur sequentially, and then report the XOR of the two inputs. This task requires memory as well as some computation.

### Imports and housekeeping

```python
# - Disable warnings
import warnings
warnings.filterwarnings('ignore')
from IPython.display import clear_output

# - General imports and configuration
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('default')
plt.rcParams['figure.figsize'] = [12, 4]
plt.rcParams['figure.dpi'] = 300

from typing import Dict, Tuple, Any, Callable, Union, List, Optional
import os
import json

# - Import rate and ADS layers from Rockpool
from rockpool.layers import RecRateEulerJax_IO, H_tanh
from rockpool.networks import NetworkADS
from rockpool.timeseries import TSContinuous

# - Jax speeds up processing
from jax import jit
import jax.numpy as jnp
```

Now we generate some helper functions to define the task.

```python
def generate_xor_sample(total_duration, dt, amplitude=1, use_smooth=True, plot=False):
    """
    Generates a temporal XOR signal
    """
    input_duration = 2/3*total_duration
    
    # Create a time base
    t = np.linspace(0,total_duration, int(total_duration/dt)+1)

    # - Generate a random input signal with two bumps and variable durations
    first_duration = np.random.uniform(low=input_duration/10, high=input_duration/4 )
    second_duration = np.random.uniform(low=input_duration/10, high=input_duration/4 )
    end_first = np.random.uniform(low=first_duration, high=2/3*input_duration-second_duration)
    start_first = end_first - first_duration
    start_second = np.random.uniform(low=end_first + 0.1, high=2/3*input_duration-second_duration) # At least 200 ms break
    end_second = start_second+second_duration
    data = np.zeros(int(total_duration/dt)+1)
    
    # - Random input amplitudes
    i1 = np.random.rand() > 0.5
    i2 = np.random.rand() > 0.5
    response = i1 ^ i2
    a1 = 1 if i1 else -1
    a2 = 1 if i2 else -1
    
    # - Make an input label
    input_label = 0
    if i1 and i2:
        input_label = 0
    elif i1 and not i2:
        input_label = 1
    elif not i1 and i2:
        input_label = 2
    else:
        input_label = 3
    
    # - Generat the temporal signals
    data[(start_first <= t) & (t < end_first)] = a1
    data[(start_second <= t) & (t < end_second)] = a2

    # - Smooth and scale the input
    if use_smooth:
        sigma = 10
        w = (1/(sigma*np.sqrt(2*np.pi)))* np.exp(-((np.linspace(1,1000,int(1/dt))-500)**2)/(2*sigma**2))
        w = w / np.sum(w)
        data = amplitude*np.convolve(data, w, "same")
    else:
        data *= amplitude
        
    # - Generate a target signal
    target = np.zeros(int(total_duration/dt)+1)
    ar = 1.0 if response else -1.0
    target[int(1/dt*(end_second+0.05)):int(1/dt*(end_second))+int(1/dt*0.3)] = ar
    
    # - Smooth the target
    if use_smooth:
        sigma = 20
        w = (1/(sigma*np.sqrt(2*np.pi)))* np.exp(-((np.linspace(1,1000,int(1/dt))-500)**2)/(2*sigma**2))
        w = w / np.sum(w)
        target = np.convolve(target, w, "same")
        target /= np.max(np.abs(target))

    # - Plot the trial
    if plot :
        eps = 0.05
        plt.plot(t, data)
        plt.plot(t, target)
        plt.ylim([-amplitude-eps, amplitude+eps])
        plt.xlabel('Time (s)')
        plt.legend(['Input', 'Target'])
        plt.box('off')
    return (data[:int(total_duration/dt)], target[:int(total_duration/dt)], input_label)
```

We can call this function to generate and plot a quick example trial.

```python
# Quick demonstration of the input data
generate_xor_sample(total_duration=1.0, dt=0.001, plot=True);
```

```python
# - Create some helper functions

# - Moving average for smoothing the response
def filter_1d(data, alpha = 0.9):
    last = data[0]
    out = np.zeros((len(data),))
    out[0] = last
    for i in range(1,len(data)):
        out[i] = alpha*out[i-1] + (1-alpha)*data[i]
        last = data[i]
    return out

# - Function to progressively scale k (error signal amplitude)
def k_step_function(total_num_iter, step_size, start_k):
    stop_k = step_size
    num_reductions = int((start_k - stop_k) / step_size) + 1
    reduce_after = int(total_num_iter / num_reductions)
    reduction_indices = [i for i in range(1,total_num_iter) if (i % reduce_after) == 0]
    k_of_t = np.zeros(total_num_iter)
    if(total_num_iter > 0):
        k_of_t[0] = start_k
        for t in range(1,total_num_iter):
            if(t in reduction_indices):
                k_of_t[t] = k_of_t[t-1]-step_size
            else:
                k_of_t[t] = k_of_t[t-1]

    return k_of_t
```

Now we initialise some global parameters for training the rate networn, such as the activation function `activation_func`, network size `num_units`, number of training epochs `num_epochs`, etc.

```python
# - Initialize some variables
# - Define global parameters
verbose = 1 # - Level of verbosity also for the ADS Network (use 1 for this tutorial)
activation_func = H_tanh # - Should be kept to tanh
duration = 1.0 # - Duration of the signal in seconds
dt = 1e-3 # - Simulation time step in seconds
amplitude = 1.0
num_units = 64 # - Corresponds to the number of neurons in the rate network
noise_std = 0.0
num_epochs = 100
num_batches = 100
num_channels = 1 # - Number of input channels. Here 1 for the 1D input
num_targets = 1 # - Output dimension is also 1D
time_base = np.arange(0,duration,dt)
```

We can now randomly initialise a rate network, by defining input, recurrent and output weights `w_in`, `w_rec` and `w_out`, as well as biases `bias` and time constants `tau`. The recurrent weights are normalised such that the eigenspectrum radius is exactly 0.95.

```python
# - Create the rate network
w_in = 10.0 * (np.random.rand(num_channels, num_units) - .5)
w_rec = 0.2 * (np.random.rand(num_units, num_units) - .5)
w_rec -= np.eye(num_units) * w_rec
w_out = 0.4*np.random.uniform(size=(num_units, num_targets))-0.2
bias = 0.0 * (np.random.rand(num_units) - 0.5)
tau = np.linspace(0.01, 0.1, num_units)

# - Normalise recurrent weight eigenspectrum
sr = np.max(np.abs(np.linalg.eigvals(w_rec)))
w_rec = w_rec / sr * 0.95
```

The cell below either loads a pre-trained rate network, or else creates a new rate network with the random initialisation.

```python
if(not os.path.exists("./Resources")):
    print("Creating Resources folder"); os.mkdir("./Resources")

model_path = "./Resources/temporal_xor_rate_model_64.json"
loaded_pretrained = False
if(os.path.exists(model_path)):
    with open(model_path, 'rb') as f:
        loaded_net = json.load(f)
        if(loaded_net['activation_func'] == 'tanh'):
            activation_func = H_tanh
        else:
            assert(False), "Please use tanh"
        lyr_hidden = RecRateEulerJax_IO(activation_func=activation_func,
                                        w_in=loaded_net['w_in'],
                                        w_recurrent=loaded_net['w_recurrent'],
                                        w_out=loaded_net['w_out'],
                                        tau=loaded_net['tau'],
                                        bias=loaded_net['bias'],
                                        dt=loaded_net['dt'],
                                        noise_std=loaded_net['noise_std'],
                                        name="hidden")
        loaded_pretrained = True
        w_out = loaded_net['w_out']
        print("Loaded pretrained layer")

else:
    # - Create the rate layer
    lyr_hidden = RecRateEulerJax_IO(activation_func=activation_func,
                                                w_in=w_in,
                                                w_recurrent=w_rec,
                                                w_out=w_out,
                                                tau=tau,
                                                bias=bias,
                                                dt=dt,
                                                noise_std=noise_std,
                                                name="hidden")
```

The cell below contains a training loop for the rate network.

```python
if not loaded_pretrained:
    # - Start training
    for epoch in range(num_epochs):
        num_samples = 0
        mvg_avg_mse = 0

        for batch_id in range(num_batches):
            # - Generate a single temporal XOR trial
            data, target, _ = generate_xor_sample(total_duration=duration, dt=dt, amplitude=amplitude)
            ts_data = TSContinuous(time_base, data)
            ts_target = TSContinuous(time_base, target)

            # - Train the rate network against this trial
            lyr_hidden.reset_time()
            l_fcn, g_fcn, o_fcn = lyr_hidden.train_output_target(
                ts_data,
                ts_target,
                is_first = (batch_id == 0) and (epoch == 0),
                opt_params={"step_size": 1e-4},
            )

            # - Evolve the rate network
            ts_out = lyr_hidden.evolve(ts_data)

            # - Plot the trial if requested
            if verbose > 0:
                clear_output(wait=True)
                plt.clf()
                ts_target.plot(linestyle='--')
                ts_out.plot()
                plt.draw()
                plt.pause(0.0001)

            # - Compute mean-squared error against the target signal
            mse = np.linalg.norm(ts_target.samples - ts_out.samples)**2
            mvg_avg_mse = mvg_avg_mse * num_samples + mse
            num_samples += 1
            mvg_avg_mse /= num_samples

            print(f"Moving average MSE is {mvg_avg_mse}; max tau {np.max(lyr_hidden.tau)}; mean tau {np.mean(lyr_hidden.tau)}")
            print(f"max bias {np.max(lyr_hidden.bias)}; mean bias {np.mean(lyr_hidden.bias)}")

    lyr_hidden.reset_all()
    lyr_hidden.noise_std = 0.0
    
    # - Save the trained rate net
    lyr_hidden.save_layer(model_path)
    print("Saved layer")
else:
    print("No training, pretrained layer was loaded")
```

Now we can define the ADS spiking network. We need to choose the number of spiking neurons `num_neurons`, the membrane time constant `tau_mem`, the fast and slow synaptic time constants `tau_syn_fast` and `tau_slow`.

We generate a random encoder / decoder matrix `D`, which lets us analytically define the fast recurrent weights `weights_fast`.

The slow recurrent weights `weights_slow` are initialised to zero.

```python
# - Define parameters used by the ADS Network
Nc = num_units
num_neurons = 768 # - This can be varied

dt = 0.001
duration = 1.0
time_base = np.arange(0,duration,dt)
lambda_d = 20
lambda_v = 20
tau_mem = 1/ lambda_v

tau_slow = 0.07 # 70ms
tau_out = 0.07

# - See Boerlin et al 2013 for the calculations below
tau_syn_fast = tau_slow
mu = 0.0005
nu = 0.0001
D = np.random.randn(Nc, num_neurons) / Nc
weights_fast = (D.T@D + mu*lambda_d**2*np.eye(num_neurons))

# - Fill the diagonal with zeros
np.fill_diagonal(weights_fast, 0)

# - Define thresholds after Boerlin et al 2013
v_thresh = (nu * lambda_d + mu * lambda_d**2 + np.sum(abs(D.T), -1, keepdims = True)**2) / 2

# - Start with zero slow weights 
weights_slow = np.zeros((num_neurons,num_neurons))

# - Define learning algorithm parameters
eta = 0.0001
k = 100
```

Now we can rescale the weights to match a threshold of 1.0, a rest potential of 0.5.

```python
# - Calculate weight matrices for realistic neuron settings
v_thresh_target = 1.0*np.ones((num_neurons,)) # - V_thresh
v_rest_target = 0.5*np.ones((num_neurons,)) # - V_rest = b

b = v_rest_target
a = v_thresh_target - b

# - Feedforward weights: Divide each column i by the i-th threshold value and multiply by i-th value of a
D_realistic = a*np.divide(D, v_thresh.ravel())
weights_in_realistic = D_realistic
weights_out_realistic = np.copy(D_realistic).T
weights_fast_realistic = a*np.divide(weights_fast.T, v_thresh.ravel()).T # - Divide each row

# - Reset is given by v_reset_target = b - a
v_reset_target = b - a
noise_std_realistic = 0.00
```

Here we generate the ADS network instance.

```python
# - Initialize network
# - Use discretize_dynapse=True if the DYNAP-SE II constraints should be used
net = NetworkADS.SpecifyNetwork(N=num_neurons,
                                Nc=Nc,
                                Nb=num_neurons,
                                weights_in= weights_in_realistic * tau_mem,
                                weights_out= weights_out_realistic,
                                weights_fast= - weights_fast_realistic / tau_syn_fast * 3, # - We will not use the fast recurrent connections here
                                weights_slow = weights_slow,
                                eta = eta,
                                k = k,
                                noise_std = noise_std_realistic,
                                dt = dt,
                                v_thresh = v_thresh_target,
                                v_reset = v_reset_target,
                                v_rest = v_rest_target,
                                tau_mem = tau_mem,
                                tau_syn_r_fast = tau_syn_fast,
                                tau_syn_r_slow = tau_slow,
                                tau_syn_r_out = tau_out,
                                record = True,
                                discretize_dynapse = False,
                               )

amplitude = 10 / tau_mem
```

### Data loader function
This helper function passes the temporal-XOR input through the rate network and returns:

- The transformed input $\hat{\textbf{F}}\textbf{c}$ that gets fed into the spiking network
- The rate network dynamics $\hat{\textbf{x}}$ that the spiking network needs to generate
- The output of the rate network for plotting and error calculation

```python
def get_data(data):
        ts_data = TSContinuous(time_base, data)
        
        # - Pass through the rate network
        ts_rate_out = lyr_hidden.evolve(ts_data)
        lyr_hidden.reset_all()
        
        # - Get the target dynamics
        ts_rate_net_target_dynamics = lyr_hidden.res_acts_last_evolution
        
        # - Get the input into the spiking network
        ts_spiking_in = TSContinuous(lyr_hidden.res_inputs_last_evolution.times,amplitude*lyr_hidden.res_inputs_last_evolution.samples)
        
        return (ts_spiking_in, ts_rate_net_target_dynamics, ts_rate_out)
```

```python
# - Test the helper function
(ts_test_spiking_in,
 ts_test_rate_dynamics,
 ts_test_rate_out) = get_data(
    data = generate_xor_sample(total_duration=duration, dt = dt)[0]
)

plt.figure(figsize=(12, 8))
ts_test_rate_dynamics.plot(stagger = 0.5);
```

```python
# - Create learning schedules for eta and k

num_epochs = 5
num_samples_per_epoch = 1000

# - Create step schedule for k
step_size = net.lyrRes.k_initial / 8
total_num_iter = num_epochs*num_samples_per_epoch
k_of_t = k_step_function(total_num_iter=total_num_iter,
                            step_size=step_size,
                            start_k = net.lyrRes.k_initial)
if(total_num_iter > 0):
    f_k = lambda t : np.maximum(step_size,k_of_t[t])
    if(verbose > 0):
        plt.plot(np.arange(0,total_num_iter),f_k(np.arange(0,total_num_iter))); plt.title("Decay schedule for k"); plt.show()
else:
    f_k = lambda t : 0
    
# - Create schedule for eta
a_eta = net.lyrRes.eta_initial
b_eta = (total_num_iter/2) / np.log(100)
c_eta = 0.0000001
f_eta = lambda t,a_eta,b_eta : a_eta*np.exp(-t/b_eta) + c_eta

if(verbose > 0):
    plt.plot(np.arange(0,total_num_iter),f_eta(np.arange(0,total_num_iter),a_eta,b_eta))
    plt.title("Decay schedule for eta"); plt.show()
```

## Notes about training
There are a few things to keep in mind when training an ADS Network:

- During training, an error-dependent current $k\textbf{F}^T \textbf{e}$ is fed into the neurons. This current steers the dynamics into the correct direction so that they follow the target dynamics. This will lead to low reconstruction errors during training, which does not directly imply that the reconstruction error will be low during validation, when the error feedback is removed.
- The hyperparameter $k$ should be chosen in a way that the magnitudes of error- and output current match. As stated before, too large error current will lead to poor generalization performance
- Scale of the output weights: This hyperparameter is important. For smooth target dynamics a scale < 1 (e.g. 0.5) is good. For more sharp and abrupt output dynamics a scale >= 1 is good, as in this tutorial.

```python
# - Train the ADS Network
np.random.seed(42)
time_horizon = 50
recon_errors = np.ones((time_horizon,))
avg_training_acc = np.zeros((time_horizon,)); avg_training_acc[:int(time_horizon/2)] = 1.0
was_first = False

def validate():
    # - Evolve over new examples without learning
    for _ in range(10):
        net.reset_all()
        data, target, _ = generate_xor_sample(total_duration=duration, dt=dt, amplitude=1.0)
        (ts_spiking_in, _, ts_rate_out) = get_data(data=data)
        test_sim = net.evolve(ts_input = ts_spiking_in, verbose = False)
        out_val = test_sim["output_layer_0"].samples
        final_out = out_val @ lyr_hidden.w_out
        final_out = filter_1d(final_out, alpha=0.95)
        
        final_out_ts = TSContinuous(
            test_sim["output_layer_0"].times,
            final_out,
            name = 'Spiking',
        )
        clear_output(wait=True)
        
        plt.clf()
        final_out_ts.plot()
        ts_rate_out.plot()
        plt.plot(time_base, target, label="Target")
        plt.legend(['Spiking', 'Rate', 'Target'])
        
        plt.ylim([-1.5,1.5])
```

```python
# - The ADS network training loop
for epoch in range(num_epochs):
    for batch_id in range(num_samples_per_epoch):
        net.reset_all()
        
        # - Get data
        data, target, _ = generate_xor_sample(total_duration=duration, dt=dt, amplitude=1.0)
        (ts_spiking_in, ts_rate_net_target_dynamics, ts_rate_out) = get_data(data=data)
        
        # - Train one trial
        train_sim = net.train_step(ts_input = ts_spiking_in,
                                   ts_target = ts_rate_net_target_dynamics,
                                   k = f_k(epoch*num_samples_per_epoch+batch_id),
                                   eta = f_eta(epoch*num_samples_per_epoch+batch_id, a_eta, b_eta),
                                   verbose = False)
        

        # - Compute train loss & update the moving averages
        out_val = train_sim["output_layer_0"].samples
        target_val = ts_rate_net_target_dynamics.samples
        error = np.sum(np.var(target_val-out_val, axis=0, ddof=1)) / (np.sum(np.var(target_val, axis=0, ddof=1)))
        recon_errors[1:] = recon_errors[:-1]
        recon_errors[0] = error
        # - Initialize the error array with the first reconstruction error we get
        if not was_first:
            was_first = True
            recon_errors = [error for _ in range(time_horizon)]

        # - Compute the final output using the reconstructed dynamics and the trained read-out of the rate net
        final_out = out_val @ lyr_hidden.w_out
        # - ..and filter
        final_out = filter_1d(final_out, alpha=0.95)

        # - Some plotting
        if(verbose > 0):
            # clear_output(wait=True)
            plt.clf()
            plt.plot(time_base, final_out, label="Spiking")
            plt.plot(time_base, target, label="Target")
            plt.plot(time_base, ts_rate_out.samples, label="Rate")
            plt.ylim([-1.5,1.5])
            plt.legend()
            plt.draw()
            plt.pause(0.001)
        
        print(("Avg. reconstruction error %.4f" % (np.mean(recon_errors))), "K", net.lyrRes.k, "Epoch", epoch, "Batch", batch_id)
```
