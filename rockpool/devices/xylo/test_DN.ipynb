```python
# a simple code to test the results
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0, np.pi)
y=np.sin(x)

plt.plot(x,y)
plt.xlabel('time')
plt.ylabel('amplitude')
plt.grid(True)
plt.title('A harmonic waveform')

plt.show()
```
